```python
import pandas as pd
import numpy as np
!pip install unidecode
from unidecode import unidecode
```

```python
data = pd.read_excel("sample_data/tabela_veg.xlsx", sheet_name='Plan1')
```

```python
data.head()
```

```python
data_ing = pd.DataFrame(data['Ingredientes'])
data_class = pd.DataFrame(data['Classificação'])
data_name = pd.DataFrame(data['Nome do Produto'])
df = data_class.join(data_name)
df.head()

#data = data.drop(columns=['Ingredientes'])
```

```python
from pandas.core.frame import DataFrame

for i in range(278):
  aux = data_ing.at[i,'Ingredientes'].split(",")
  for w in aux:
    w = unidecode(w.lower().lstrip().replace('.',''))
    df.at[i,w] = 1

df = df.replace(np.nan,0)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import plotly.express as px

non_numeric = ['Classificação', 'Nome do Produto']
df_numeric = df.drop(non_numeric, axis=1)

m = TSNE(n_components=2,perplexity=12,learning_rate=50.0,metric='euclidean', init='random')
tsne_features = m.fit_transform(df_numeric)
tsne_features[1:4, :]

df['x'] = tsne_features[:,0]
df['y'] = tsne_features[:,1]

fig = px.scatter(df, x="x", y="y",color="Classificação")
fig.show()
```
