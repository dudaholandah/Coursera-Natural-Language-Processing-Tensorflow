```python
import pandas as pd
import numpy as np
!pip install unidecode
from unidecode import unidecode
```

```python
data = pd.read_excel("sample_data/tabela_veg.xlsx", sheet_name='Plan1')
```

```python
data.head()
```

```python
data_ing = pd.DataFrame(data['Ingredientes'])
data_class = pd.DataFrame(data['Classificação'])
data_name = pd.DataFrame(data['Nome do Produto'])
df = data_class.join(data_name)
df.head()

#data = data.drop(columns=['Ingredientes'])
```

```python
from pandas.core.frame import DataFrame

for i in range(278):
  aux = data_ing.at[i,'Ingredientes'].split(",")
  for w in aux:
    w = unidecode(w.lower().lstrip().replace('.',''))
    df.at[i,w] = 1

df = df.replace(np.nan,0)
df.head()
```
